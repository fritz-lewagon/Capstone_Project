```python
import requests
import time
import pandas as pd
import bs4
from selenium import webdriver
```

```python
base_url = 'http://startupshub.catalonia.com'

startup_list = {'name':[],
                'description':[],
                'business_model':[],
                'target':[],
                'keywords':[],
                'stage':[],
                'founded':[],
                'employees':[]}
```

```python
# This function extracts the url of the detailed start-up page and fetches additional information into the dictionary
def get_details(startup):
    
    # get url and save content in bs4 object
    page = requests.get(base_url + startup.find(class_ = 'text-comp').find('a').get('href'))
    detail = bs4.BeautifulSoup(page.text, "html.parser")

    try:
        startup_list['stage'].append(detail.find(class_ = 'stage information-item').find('strong').attrs['value'])
    except:
        startup_list['stage'].append('')
    try:
        startup_list['founded'].append(detail.find(class_ = 'founded information-item').get_text().strip())
    except:
            startup_list['founded'].append('')
    try:
        startup_list['employees'].append(detail.find(class_ = 'employers information-item').find('strong').attrs['value'])
    except:
            startup_list['employees'].append('')
```

```python
driver_path = 'C:/Users/maxig/OneDrive/Desktop/Capstone Project/Chromedriver.exe'
driver = webdriver.Chrome()
driver.get('http://startupshub.catalonia.com/list-of-startups')

startups = bs4.BeautifulSoup(driver.page_source).find_all(class_ = 'activitat')
```

```python
for startup in startups:
    
    try:
        startup_list['name'].append(startup.find(class_ = 'titol').get_text().strip())            
    except:
        startup_list['name'].append('')
        
    try:
        startup_list['description'].append(startup.find(class_ = 'description').get_text().strip())
    except:
        startup_list['description'].append('')
        
    try:
        startup_list['business_model'].append(startup.find(class_ = 'b-model').get_text().replace('\t','').replace('\n','').replace('\r','').strip())
    except:
        startup_list['business_model'].append('')
        
    try:
        startup_list['target'].append(startup.find(class_ = 'target').get_text().replace('\t','').replace('\n','').replace('\r','').strip())
    except:
        startup_list['target'].append('')
        
    try:
        startup_list['keywords'].append(startup.find(class_ = 'cats').get_text().replace('\t','').replace('\n','').replace('\r','').strip())
    except:
        startup_list['keywords'].append('')
        
    get_details(startup)
        
        
    
```

```python
while (True):
    time.sleep(3)
    try:
        driver.find_elements_by_class_name('next-1')[0].click()
        startups = bs4.BeautifulSoup(driver.page_source).find_all(class_ = 'activitat')
        
        for startup in startups:
            try:
                startup_list['name'].append(startup.find(class_ = 'titol').get_text().strip())
            except:
                startup_list['name'].append('')
            try:
                startup_list['description'].append(startup.find(class_ = 'description').get_text().strip())
            except:
                startup_list['description'].append('') 
            try:
                startup_list['business_model'].append(startup.find(class_ = 'b-model').get_text().replace('\t','').replace('\n','').replace('\r','').strip())
            except:
                startup_list['business_model'].append('')
            try:
                startup_list['target'].append(startup.find(class_ = 'target').get_text().replace('\t','').replace('\n','').replace('\r','').strip())
            except:
                startup_list['target'].append('')
            try:
                startup_list['keywords'].append(startup.find(class_ = 'cats').get_text().replace('\t','').replace('\n','').replace('\r','').strip())
            except:
                startup_list['keywords'].append('')
        
            get_details(startup)
            
    except:
        break
                
                
                
```

```python
df = pd.DataFrame(startup_list)
```
