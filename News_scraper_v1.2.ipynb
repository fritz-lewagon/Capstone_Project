```python
import bs4
import requests
import numpy as np
import pandas as pd
```

```python
def get_page(url):
    
    result = requests.get(url)
    soup = bs4.BeautifulSoup(result.text, 'html.parser')
    
    return soup
```

```python
url = "https://elreferente.es/"
soup = get_page(url)

# different article elements (a1 = top article, a2 = sub articles, a3 = related articles)
a1 = soup.find_all("h2", class_="news-item-title")
a2 = soup.find_all("h4", class_="news-item-title")
a3 = soup.find_all("div", class_="rel-news-title")
```

```python
all_articles = a1 + a2 + a3
```

```python
def get_title(article_object):
    return article_object.text.replace("\t", "").replace("\n", "").replace('"', '').replace('  ','')
```

```python
get_title(a1[0])
```

```python
def get_url(article_object):
    return article_object.find('a')['href']
```

```python
get_url(a3[0])
```

```python
def get_article_text(article_object):
    
    article = requests.get(get_url(article_object))
    soup_content = bs4.BeautifulSoup(article.content, 'html.parser')
    
    body = soup_content.find_all('div', class_="news-item-content p-with-margin blue-links")
    try:
        x = body[0].find_all('p')
        list_paragraphs = []
        
        for p in np.arange(0, len(x)):
            paragraph = x[p].get_text()
            list_paragraphs.append(paragraph)
            final_article = " ".join(list_paragraphs)
            final_article = final_article.replace('\xa0', '')
    except IndexError:
        final_article = None

    return final_article
```

```python
get_article_text(a3[0])
```

```python
def get_article_date(article_object):
    
    import datetime

    article = requests.get(get_url(article_object))
    soup_content = bs4.BeautifulSoup(article.content, 'html.parser')
    
    try:
        day = int(soup_content.find('div', class_='date-day').text.replace('\n', '').replace(' ',''))
        month = soup_content.find('div', class_='date-month').text.replace('\n', '').replace(' ','').replace('\t', '')
        year = int(soup_content.find('div', class_='date-year').text.replace('\n', '').replace(' ',''))
        
        m_all = {'ene': 1,
             'feb': 2,
             'mar': 3,
             'abr': 4,
             'may': 5,
             'jun': 6,
             'jul': 7,
             'ago': 8,
             'sep': 9,
             'oct': 10,
             'nov': 11,
             'dic': 12
            }
        date1 = datetime.date(year, m_all[month], day)
    
    except:
        date1 = None
    
    return date1
    
```

```python
def get_article_tags(article_object):
    
    article = requests.get(get_url(article_object))
    soup_content = bs4.BeautifulSoup(article.content, 'html.parser')
    
    tags = []

    for i in soup_content.find_all('div', class_='single-item-tag'):
        tags.append(i.text.replace(' ','').replace('\n', ''))
    
    # convert tags list to string (maybe more convenient to have list?)
    string = ', '.join([str(i) for i in tags])
    
    return string
```

```python
def get_news_category(article_object):
    article = requests.get(get_url(article_object))
    soup_content = bs4.BeautifulSoup(article.content, 'html.parser')
    try:
        category1 = soup_content.find('div', class_="news-item-section").text.replace(' ','').replace('\n', '')
    except:
        category1 = None
    
    return category1
```

```python
def comp_urls(article_object):
    article = requests.get(get_url(article_object))
    soup_content = bs4.BeautifulSoup(article.content)

    urls = []

    for i in soup_content.find_all('h5', class_='startup-item-title'):
        urls.append(i.find('a')['href'])

    return urls
```

```python
def get_comp_name(comp_url):
    page = requests.get(comp_url)
    page_content = bs4.BeautifulSoup(page.content)
    name = page_content.find('h2', class_ = 'title').text.replace('\n', '').replace('  ', '')

    return name
    
```

```python
# basic infos

def basic_infos(comp_url):
    
    page = requests.get(comp_url)
    page_content = bs4.BeautifulSoup(page.content) 

    infos = []
    for i in page_content.find('div', class_='flex-row info-container').find_all('div', class_='value'):
        infos.append(i.text.replace(' ', '').replace('\n', ''))

    # founding date
    date = infos[0]
    
    # reformatting of total_funding
    if infos[1] == 'N/D':
        invest = ''
    else:
        invest = float(infos[1].replace(',', '.').replace('M€', ''))
    
    # number of investors
    numinvest = infos[2]
    
    return [date, invest, numinvest]
```

```python
def det_infos(comp_url):
    
    page = requests.get(comp_url)
    page_content = bs4.BeautifulSoup(page.content)
    
    # employees
    empl = page_content.find('div', class_ = 'value blur-no-auth').text.replace('\n', '').replace(' ', '')
    if len(empl)>1:
        empl = empl
    else:
        empl = ''
        
    try:
        loca = page_content.find('div', class_ = 'value icon-location').text.replace('\n', '').replace('  ', '')
    except:
        loca = ''
    
    try:
        webs = page_content.find('a', class_ = 'blue')['href']
    except:
        webs = ''
        
    try:
        sect = page_content.find('div', class_ = 'flex-row tags highlight').text.replace('\n', '').replace(' ', '')
    except:
        sect = ''
    
    try:
        cust = page_content.find('div', class_ = 'section section-info blue-links').find_all('div', class_= 'value')[-1].text.replace('\n', '').replace(' ', '')
    except:
        cust = ''
    
    try:
        desc = page_content.find('div', class_ = 'description blue-links').text.replace('\n', '').replace('\xa0', '')
    except:
        desc = ''
        
    return [empl, loca, webs, sect, cust, desc]
    
```

```python
## company dataframe

comp_name = []

# basic info
date = []
invest = []
numinvest = []

# comp detailed infos
empl = []
loca = []
webs = []
sect = []
cust = []
desc = []
bm = []
stage = []


for i in all_articles:

    if len(comp_urls(i)) > 0:
        
        for j in comp_urls(i):
            comp_name.append(get_comp_name(j))
            date.append(basic_infos(j)[0])
            invest.append(basic_infos(j)[1])
            numinvest.append(basic_infos(j)[2])
            #empl.append(det_infos(j)[0]) -> range different to Maxi
            empl.append('')
            loca.append(det_infos(j)[1])
            webs.append(det_infos(j)[2])
            sect.append(det_infos(j)[3])
            cust.append(det_infos(j)[4])
            desc.append(det_infos(j)[5])
            bm.append('')
            stage.append('')
    
elr_comps = pd.DataFrame(data = {'comp_name': comp_name,
                                 'description': desc,
                                 'business_model': bm,
                                 'customer': cust,
                                 'keywords': sect,
                                 'stage': stage,
                                 'total_funding': invest,
                                 'num_investors': numinvest,
                                 'date_founded': date,
                                 'location': loca,
                                 'employees': empl,
                                 'website': webs})
```
